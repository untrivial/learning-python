```python
import pandas as pd
# the object type here is Dataframe

df = pd.read_csv('pokemon_data.csv')

df_xlsx = pd.read_excel('pokemon_data.xlsx')

df_txt = pd.read_csv('pokemon_data.txt', delimiter='\t') # tab-separated file

# print(df) <- all data
print(df.head(3)) # top 3 rows
print(df.tail(3)) # bottom 3 rows
```

```python
## Read Headers
print(df.columns)

## Read each Column
print(df['Name'][0:3]) # top 5
print(df[['Name', 'Type 1', 'HP']][0:3]) # multiple columns

## Read Each Row
print(df.iloc[1]) # iloc means integer location, here it pull 1 row
print(df.iloc[1:4])

for index, row in df.iterrows():
    print(index, row['Name'])
    
print(df.loc[df['Type 1'] == "Fire"])

## Read a specific location (Row, Column)
print(df.iloc[2,1]) # 2nd Row, 1st Column
```

```python
df.describe() # statistical metrics of data
df.sort_values('Name')
df.sort_values('Name', ascending=False)
df.sort_values(['Type 1', 'HP'], ascending=[True, False]) # sort by Type 1 (ascending), then by HP (descending)
```

```python
# new column that sums all the stats
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```
